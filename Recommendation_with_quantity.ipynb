```python
import pandas as pd

df = pd.read_csv('combined.csv')

df.head()
```

```python
import pandas as pd
import numpy as np

weights = [0.6, 0.3, 0.1]

# Generate random quantities based on the weights
random_quantities = np.random.choice([1, 2, 3], size=len(df), p=weights)

# Add the new 'quantity' column to the DataFrame
df['quantity'] = random_quantities


df.head()
```

```python
df[:20]
```

```python
# Create a pivot table
pivot_df = df.pivot_table(index='asin', columns='orderNumber', values='quantity', aggfunc='sum', fill_value=0)
```

```python
pivot_df
```

```python
pip install scikit-learn
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarities = cosine_similarity(pivot_df)

# Find the indices of the ASINs you want to compare
asin1_index = pivot_df.index.get_loc("B00062NHH0")
asin2_index = pivot_df.index.get_loc("B0029Y2KNA")

# Get the cosine similarity value
similarity = cosine_similarities[asin1_index, asin2_index]

print(f"Cosine Similarity between B00062NHH0 and B0029Y2KNA: {similarity}")
```

```python
# Create a DataFrame for the cosine similarity matrix
asin_similarity_df = pd.DataFrame(cosine_similarities, index=pivot_df.index, columns=pivot_df.index)
```

```python
asin_similarity_df
```

```python
def recommend_top_similar_asins(target_asin, similarity_df, top_n=10):
    # Get the row corresponding to the target ASIN
    target_row = similarity_df.loc[target_asin]

    # Sort the row in descending order and get top similar ASIN indices
    similar_asin_indices = target_row.sort_values(ascending=False).index[1:top_n+1]

    # Get the similarity values for the top similar ASINs
    similar_asin_values = target_row[similar_asin_indices]

    # Create a DataFrame for the recommendations
    recommendations_df = pd.DataFrame({
        'Recommended ASIN': similar_asin_indices,
        'Similarity Value': similar_asin_values
    })

    return recommendations_df
```

```python
target_asin = "B00007GDFV"

# Get recommendations
recommendations = recommend_top_similar_asins(target_asin, asin_similarity_df)
print(recommendations)
```

```python
import matplotlib.pyplot as plt

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.barh(recommendations['Recommended ASIN'], recommendations['Similarity Value'], color='skyblue')
plt.xlabel('Similarity Value')
plt.ylabel('Recommended ASIN')
plt.title(f'Top Recommended ASINs for {target_asin} and Their Similarity Values')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest similarity on top
plt.show()
```

```python
import pandas as pd

asin_df = pd.read_csv('asin_similarity.csv', index_col=0)

asin_df.head()
```

```python
import pandas as pd

n = 10  # Number of top pairs to find
top_similar_pairs = []

# Loop through each row to find the top pairs
for row_idx, row in asin_df.iterrows():
    for col_idx, similarity in row.items():
        if row_idx != col_idx:  # Exclude pairs with the same ASIN
            top_similar_pairs.append((row_idx, col_idx, similarity))

# Sort the pairs by similarity score in descending order
top_similar_pairs.sort(key=lambda x: x[2], reverse=True)

# Display the top 10 similar pairs
for pair in top_similar_pairs[:n]:
    print(pair)
```

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt

pairs = [f"{pair[0]} - {pair[1]}" for pair in top_similar_pairs]
scores = [pair[2] for pair in top_similar_pairs]

plt.figure(figsize=(10, 6))
plt.barh(pairs, scores, color='skyblue')
plt.xlabel('Similarity Score')
plt.ylabel('ASIN Pairs')
plt.title('Top ASIN Pairs by Similarity Score')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest scores at the top
plt.tight_layout()
plt.show()
```
